```python
#	program to find next character given a characters using 
#	simple rnn by Santhosh on 19-7-18

#preparation of dataset

file=open('vocab.txt')
lines=list()
for line in file:
	line=line.strip().lower()
	line.encode("ascii","ignore")
	if len(line)==0:
		continue
	lines.append(line)
file.close()
text=" ".join(lines)

#generating a characters vocabulary,
chars=set([c for c in text]) 
nb_chars=len(chars)
char2index=	dict((c,i) for i,c in enumerate(chars))
index2char=	dict((i,c) for i,c in enumerate(chars))
print(nb_chars)

STEP = 1
input_chars = []
label_chars = []
for i in range(0, len(text) - 1, STEP):
	input_chars.append(text[i])
	label_chars.append(text[i + 1])

#preparing training data
import numpy as np
x=np.zeros((len(input_chars),1,nb_chars),dtype=np.bool)
y=np.zeros((len(input_chars),nb_chars),dtype=np.bool)
for i,input_char in enumerate(input_chars):
	x[i,0,char2index[input_char]]=1
	y[i,char2index[label_chars[i]]]=1

#preparing the model
from keras.layers import Dense, Activation
from keras.layers.recurrent import SimpleRNN
from keras.models import Sequential
HIDDEN_SIZE = 1024


BATCH_SIZE = 128
NUM_ITERATIONS = 40
NUM_EPOCHS_PER_ITERATION = 1
NUM_PREDS_PER_EPOCH = 100

model = Sequential()
model.add(SimpleRNN(HIDDEN_SIZE, return_sequences=False,
input_shape=(1,nb_chars)))
model.add(Dense(nb_chars))
model.add(Activation("softmax"))
model.compile(loss="categorical_crossentropy", optimizer="rmsprop")
model.summary()


for iteration in range(NUM_ITERATIONS):
    print("=" * 50)
    print("Iteration #: %d" % (iteration))
    model.fit(x, y, batch_size=BATCH_SIZE, epochs=NUM_EPOCHS_PER_ITERATION)
    test_idx = np.random.randint(len(input_chars))
    test_chars = input_chars[test_idx]
    print("Generating from seed: %s" % (test_chars))
    print(test_chars, end="")
    for i in range(NUM_PREDS_PER_EPOCH):
        Xtest = np.zeros((1, 1, nb_chars))
        Xtest[0, 0, char2index[test_chars]] = 1
        pred = model.predict(Xtest, verbose=0)[0]
        ypred = index2char[np.argmax(pred)]
        print('(',test_chars,',',ypred, end="),")
        # move forward with test_chars + ypred
        test_chars =  ypred
print()
```
