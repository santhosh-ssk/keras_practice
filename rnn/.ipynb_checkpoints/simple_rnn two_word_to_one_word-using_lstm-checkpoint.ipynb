```python
#	program to find next character given a characters using 
#	simple rnn by Santhosh on 19-7-18

#preparation of dataset

file=open('story.txt')
lines=list()
for line in file:
	line=line.strip().lower()
	line.encode("ascii","ignore")
	if len(line)==0:
		continue
	lines.append(line)
file.close()
text=" ".join(lines)

#generating a words vocabulary,
words=set([c for c in text.split()]) 
nb_words=len(words)

word2index=	dict((c,i) for i,c in enumerate(words))
index2word=	dict((i,c) for i,c in enumerate(words))
print(nb_words)


input_words = []
label_words = []
words_list=text.split()
for i in range(0, len(words_list) - 2):
	input_words.append(words_list[i:i+2])
	label_words.append(words_list[i+2])

#preparing training data
import numpy as np
x=np.zeros((len(input_words),2,nb_words),dtype=np.bool)
y=np.zeros((len(input_words),nb_words),dtype=np.bool)
for i,input_word in enumerate(input_words):
    x[i,0,word2index[input_word[0]]]=1
    x[i,1,word2index[input_word[1]]]=1
    y[i,word2index[label_words[i]]]=1
    
#preparing the model
from keras.layers import Dense, Activation
from keras.layers.recurrent import LSTM
from keras.models import Sequential
HIDDEN_SIZE = 128

BATCH_SIZE = 128
NUM_ITERATIONS = 5
NUM_EPOCHS_PER_ITERATION = 50
NUM_PREDS_PER_EPOCH = 25

model = Sequential()
model.add(LSTM(HIDDEN_SIZE, return_sequences=False,
input_shape=(2,nb_words,)))
model.add(Dense(nb_words))
model.add(Activation("softmax"))
model.compile(loss="categorical_crossentropy", optimizer="rmsprop")
model.summary()

for iteration in range(NUM_ITERATIONS):
    print("=" * 50)
    print("Iteration #: %d" % (iteration))
    model.fit(x, y, batch_size=BATCH_SIZE, epochs=NUM_EPOCHS_PER_ITERATION)
    test_idx = np.random.randint(len(input_words))
    test_word = input_words[test_idx]
    print("Generating from seed: %s" % (test_word))
    for i in range(NUM_PREDS_PER_EPOCH):
        Xtest = np.zeros((1,2, nb_words))
        Xtest[0, 0, word2index[test_word[0]]] = 1
        Xtest[0, 1, word2index[test_word[1]]] = 1
        pred = model.predict(Xtest, verbose=0)[0]
        ypred = index2word[np.argmax(pred)]
        print(ypred, end=",")
        # move forward with test_chars + ypred
        test_word =[test_word[1],ypred]
print()
```
